```python
# Linear regression

import seaborn as sns
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
```

```python
# Load datastet

df = sns.load_dataset('tips')
print(df)
```

```python
X = df['total_bill'].values.reshape(-1, 1)
# print(X)
y = df['tip'].values


# Build relationship b/w x and y

model = LinearRegression()
model.fit(X, y)


# Predict values

residuals = model.predict(X) - y
sm.qqplot(residuals, line = 's')
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sample data

X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 5, 6, 8])

# Fit model

model = LinearRegression()
model.fit(X.reshape(-1, 1), y)

# Plot

plt.scatter(X, y, color = 'blue', label = 'Data')
plt.plot(X, model.predict(X.reshape(-1, 1)), color = 'red', label = 'Regression Line')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
# Linear Regression model predict house prices
#numpy, matplotlib, LinearRegression


# Define dataset

square_feet = np.array([1000, 1500, 2000, 2500, 3000])
house_prices = np.array([200000, 250000, 300000, 350000, 400000])
```

```python
# Build/Train model

model = LinearRegression() # Initialize
model.fit(square_feet.reshape(-1, 1), house_prices) # Train
```

```python
# Predict

predicted_price = model.predict(np.array([2000]).reshape(-1, 1)) # Test input
print(f"The price of 2000 sq. ft is: {predicted_price}")
```

```python
plt.scatter(square_feet, house_prices, color = 'blue', label = 'Actual Data')
plt.plot(square_feet, model.predict(square_feet.reshape(-1, 1)), color = 'red', label = 'Regression Line')
plt.xlabel('Square Feet')
plt.ylabel('House Prices')
plt.title('House Price Prediction')
plt.legend(loc = 'best')
plt.show()
```

```python
# Task 1
# nupmy, matplotlib

def linear_function(x, m = 2, c = 3):
  y = (m * x) + c
  return y
```

```python
# Task 2

values = np.linspace(-10, 10, 100)
y_values = linear_function(values)
```

```python
# Task 3

plt.plot(values, y_values, label = 'Linear (y = 2x + 3)', color = 'blue', linestyle = '-')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc = 'best')
plt.grid()
plt.show()
```
