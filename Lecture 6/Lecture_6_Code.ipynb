```python
# Create DataFrame from dictionary

import pandas as pd
import numpy as np
```

```python
employees = {
    'Emp_ID' : [101, 102, 103, 104, 105],
    'Name' : ['Hamza', 'Talha', 'Jawad', 'Honda', 'Yamaha'],
    'Age' : [24, 25, 30, np.nan, 29],
    'Salary' : [50000, 60000, 75000, 80000, np.nan],
    'Department' : ['HR', 'IT', 'Finance', 'IT', 'HR'],
    'Location' : ['New York', 'Islamabad', 'Karachi', 'Hyderabad', 'Lahore']
}

df = pd.DataFrame(employees)
print(df)
```

```python
# Missing Data

df['Age'].fillna(df['Age'].mean(), inplace = True)
#df['Salary'].fillna(df['Salary'].mean(), inplace = True)

df.dropna(subset = ['Salary'], inplace = True)

print(df)
```

```python
# Filter employees IT department

it_employees = df[df['Department'] == 'IT']
print("IT employees are:\n", it_employees)

#Filter employees Finance Department


fin_employees = df[df['Department'] == 'Finance']
print("\nFinance employees are:\n", fin_employees)
```

```python
# Sorting by Salary

df['Experience'] = np.where(df['Salary'] > 70000, 'Senior', 'Junior')
sorted_df = df.sort_values(by = 'Salary', ascending = False)
print(sorted_df)
```

```python
# New dataframe for merging

new_hires = {
    'Emp_ID' : [106, 107],
    'Name' : ['Frank', 'John'],
    'Age' : [44, 30],
    'Salary' : [55000, 65000],
    'Department' : ['HR', 'Finance'],
    'Location' : ['Berlin', 'Paris']
}

df_newhires = pd.DataFrame(new_hires)
print(df_newhires)
```

```python
df_final = pd.concat([sorted_df, df_newhires], ignore_index = True)
df_final['Experience'] = np.where(df_final['Salary'] > 70000, 'Senior', 'Junior')
print(df_final)
```

```python
x = np.linspace(0, 20, 20)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize = (10, 6))

plt.plot(x, y1, color = 'red', linestyle = '-', marker = 'o', markerfacecolor = 'yellow', label = 'Sin(x)')
plt.plot(x, y2, color = 'blue', linestyle = '--', marker = 'x', label = 'Cos(x)')
plt.xlabel('X-axis', fontsize = 12)
plt.ylabel('Y-axis', fontsize = 12)
plt.title('Sin(x) and Cos(x)', fontsize = 14)
plt.legend(loc = 'best')
plt.grid(True)
plt.annotate('Peak', xy = (1.57, 1), xytext = (3, 1.2), arrowprops = {'facecolor': 'purple', 'shrink': 0.05})
plt.show()
```

```python
# Scatter Plot

x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Some Random Dots')
plt.show()
```

```python
categories = np.random.randint(0, 5, 50)
colors = ['red', 'green', 'blue', 'purple', 'orange']

plt.scatter(x, y, c = [colors[i] for i in categories], marker = 'o')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Some Still Random But Slightly Colourful Dots')
plt.show()
```

```python
markers = ['o', 's', 'D']

for i, marker in enumerate(markers):
  plt.scatter(x[categories == i], y[categories == i], marker = marker, label = f'Category {i}', alpha = 0.1)

plt.legend(loc = 'best')
plt.xlabel('Za X-axis')
plt.ylabel('Za Y-axis')
plt.title('Bunch of random dots')
plt.show()
```

```python
z = np.random.rand(50) * 100 # Variable for scaling

plt.scatter(x, y, s = z, alpha = 0.5, c = z, cmap = 'coolwarm')
plt.colorbar(label = 'Summins gettin hot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bunch of dots showing it hot (and cold)')
plt.show()
```

```python
# Generate a linar relationship

x = np.linspace(0, 10, 50)
y = x * 3 + np.random.randn(50) * 5 # y = 3x + noise
plt.scatter(x, y, color = 'purple', label = 'Data and Stuff')


# Fit a linear trend line
m, b = np.polyfit(x, y, 1) # First degree polynomial fit
plt.plot(x, m * x + b, color = 'red', label = 'Trend Line')
plt.title('Bunch of dots with trend line')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(loc = 'best')
plt.show()
```
